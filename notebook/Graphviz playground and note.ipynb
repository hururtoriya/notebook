# Graphviz PlayGrounds


```python
%pylab inline --no-import-all
import pygraphviz as pgv
from IPython.display import Image as Image
import numpy as np
from numpy.random import *


A = pgv.AGraph() # Create Graph
n = 10
path="./assets/"
for i in range(n):
    node_a = randint(n)
    node_b = randint(n)+1
    A.add_node(node_a) 
    A.add_node(node_b)
    # We can userriibute handring after pygraphviz-0.36
    A.add_edge(node_a, node_b, label=randint(10)+1)

A.layout(prog='neato')
A.draw(path+'graph.png')
A.draw(path+'graph.eps') # We can saved eps format. 

Image(filename=path+'graph.png')
```
